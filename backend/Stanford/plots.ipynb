## Feed-Forward Neural Network

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
```

```python
features_df = pd.read_csv('features.csv')

correlation_matrix = features_df.drop(columns=['Breed']).corr()

pairs = correlation_matrix.abs().unstack().sort_values(kind="quicksort", ascending=False)
pairs = pairs[pairs < 1]
top_pairs = pairs.drop_duplicates().head(15)

top_features = list([pair[0] for pair in top_pairs.index] + [pair[1] for pair in top_pairs.index])

top_features, top_pairs

filtered_corr_matrix = correlation_matrix.loc[top_features, top_features]

plt.figure(figsize=(20, 15))
sns.heatmap(filtered_corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Heatmap of Top Correlated Features')
plt.show()
```

```python
subset_features = [
    'Feature 3', 'Feature 4', 'Feature 8', 'Feature 9', 
    'Feature 38', 'Feature 39', 'Feature 70', 'Feature 72'
]

sns.pairplot(features_df[subset_features])
plt.suptitle('Pairplot of Selected Top Correlated Features', y=1.02)
plt.show()
```
