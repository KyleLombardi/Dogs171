```python
import pandas as pd

# Load the dataset
df = pd.read_csv('features.csv')

# Extract features and labels
x = df.drop(columns=['Breed'])
y = df['Breed']

# Number of classes (breeds)
num_classes = 120

# Display the first few rows of the DataFrame
display(df.head())

# Print the number of breeds
print(f"Number of Breeds: {num_classes}")

# Create a hashmap for breed names
breed_map = {i: breed for i, breed in enumerate(y.unique(), 0)}

# Print the breed map
print("Breed Map:")
for key, value in breed_map.items():
    print(f"{key}: {value}")

# Example usage:
# Get the breed corresponding to the 31st entry in df['Breed']
index = 30
print(f"\nBreed at index {index}: {breed_map[index]}")
```

```python
# Define the filename for the output Python file
output_file = 'breed_map.py'

# Write the breed map to the output file
with open(output_file, 'w') as f:
    f.write('breed_map = {\n')
    for key, value in breed_map.items():
        f.write(f"    {key}: '{value}',\n")
    f.write('}\n')

print(f"Breeds map has been written to {output_file}.")
```
